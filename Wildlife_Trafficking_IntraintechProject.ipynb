**Mosaic plot displaying the class vs Order of different Species**

```python
def find_largest_number(numbers):
    if not numbers:
        return None  # Return None if the list is empty
    max_num = numbers[0]  # Assume the first number is the largest
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num

# Example usage:
numbers_list = [3, 7, 2, 9, 5]
largest_number = find_largest_number(numbers_list)
print("The largest number in the list is:", largest_number)
```
